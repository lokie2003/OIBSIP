```python
z = round(data.groupby(["Car_Name"])["Present_Price"].agg(['mean']),2).T
z
```

```python
data = data.merge(z.T,how='left',on="Car_Name")
```

```python
bins = [0,10000,20000,40000]
cars_bin = ["Budget","Medium","Highend"]
data["CarsRange"] = pd.cut(data['mean'],bins,right=False,labels=cars_bin)
data.head()
```

```python
newdata=data[['Year','Selling_Price','Present_Price','Driven_kms','Fuel_Type','Selling_type','Transmission','Owner','mean']]
newdata
```

```python
# encoding "Fuel_Type" Column
newdata.replace({'Fuel_Type':{'Petrol':0,'Diesel':1,'CNG':2}},inplace=True)

# encoding "Seller_Type" Column
newdata.replace({'Selling_type':{'Dealer':0,'Individual':1}},inplace=True)

# encoding "Transmission" Column
newdata.replace({'Transmission':{'Manual':0,'Automatic':1}},inplace=True)
```

```python
newdata
```

```python
#CREATING DUMMIES VARIABLES FOR ALL THE CATEGORICAL FEATURES
newdata = pd.get_dummies(columns=['Fuel_Type','Selling_type','Transmission','Owner'],data = newdata)
```

```python
newdata
```

```python
#AS WE KNOW THAT OUR DATASET HAS SOME OUTLIERS SO WE DO SCALINNG OF NUMERIC DATA
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
num_cols = ['Selling_Price','Present_Price','Driven_kms','mean']
newdata[num_cols] = scalar.fit_transform(newdata[num_cols])
```

```python
newdata
```

```python
#SEGREGATING DATA INTO INDEPENDENT AND DEPENDENT FEATURES
x = newdata.drop(columns = ["Selling_Price"])
y = newdata["Selling_Price"]
```

```python
#SPLITTING TRAIN AND TEST
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
#MODEL TRAINING BY USING RANDOM FOREST
from sklearn.ensemble import RandomForestRegressor
random_forest = RandomForestRegressor()
```

```python
#TO FIT THE MODEL
random_forest.fit(x_train,y_train)
```

```python
#PREDICT THE VALUE
y_pred = random_forest.predict(x_test)
```

```python
#MODEL EVALUATION
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred))
```
